```python
from sklearn.model_selection import train_test_split
```

```python
X = runs_selectedfeature[['horse_age', 'horse_rating', 'declared_weight', 'actual_weight', 'draw','win_odds']]
```

```python
y = runs_selectedfeature['won']
```

```python
X.describe()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

# Using KNN

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=2)
```

```python
knn.fit(X_train,y_train)
```

```python
pred = knn.predict(X_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,pred))
```

```python
print(classification_report(y_test,pred))
```

```python
error_rate = []

# Will take some time
for i in range(1,40):
    
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train,y_train)
    pred_i = knn.predict(X_test)
    error_rate.append(np.mean(pred_i != y_test))
```

```python
plt.figure(figsize=(10,6))
plt.plot(range(1,40),error_rate,color='blue', linestyle='dashed', marker='o',
         markerfacecolor='red', markersize=10)
plt.title('Error Rate vs. K Value')
plt.xlabel('K')
plt.ylabel('Error Rate')
```

```python
# NOW WITH K=10
knn = KNeighborsClassifier(n_neighbors=10)

knn.fit(X_train,y_train)
pred = knn.predict(X_test)

print('WITH K=10')
print('\n')
print(confusion_matrix(y_test,pred))
print('\n')
print(classification_report(y_test,pred))
```

# Import Keras and Package

```python
import keras

from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout
from keras.utils import to_categorical
```

```python
def classification_model():
    
    # create model
    model = Sequential()
    model.add(Dense(6, input_dim=6, activation='relu'))
    model.add(Dense(1, activation='sigmoid'))    
    # compile model
    model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
    return model
```

```python
print(X_train.shape)
```

```python
# build the model
model = classification_model()

# fit the model
model.fit(x = X_train,y = y_train, epochs=10, batch_size=1)

# evaluate the model
scores = model.evaluate(X_test, y_test, verbose=0)
```

```python
scores
```

```python
y_predictions = model.predict_classes(X_test)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_predictions))
```
